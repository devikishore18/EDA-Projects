# HISTPLOTS

# LINE PLOT

```python
plt.plot(df['Meetings_per_Week'],df['Employee_ID'])
plt.xlabel('Meetings_per_Week')
plt.ylabel('Employee_ID')
plt.title('how employees are engaged in meetings')     
plt.show()
```

```python
sns.lineplot(x=df['Job_Satisfaction'],y=df['Productivity_Score'],data=df,label='productivity score based on job satisfaction')
plt.show()
```

```python
sns.lineplot(x='Department',y='Job_Satisfaction',data=df,color='green')
plt.title('jobsatisfaction by departments')
plt.show()
```

```python
sns.lineplot(x=df['Job_Satisfaction'],y=df['Annual_Salary'],data=df,label='productivity score based on annual salary',color='red')
plt.show()
```

# BAR PLOT

```python
sns.barplot(x='Job_Satisfaction',y='Productivity_Score',data=df)
plt.show()
```

# SCATTER PLOT

```python
plt.scatter(df['Age'],df['Tasks_Completed_Per_Day'])
plt.xlabel('joblevel')
plt.ylabel('tasks completed per day')
plt.show()
```

# PIE CHARTS

```python
dept_counts = df['Department'].value_counts()
p=df['Productivity_Score'].value_counts()
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('DEPARTMENTS')
plt.show()
```

```python
overtime_counts = df['Overtime_Hours_Per_Week'].value_counts()
plt.pie(overtime_counts, labels=overtime_counts.index, autopct='%1.1f%%', 
        startangle=90)
plt.title('Overtime Distribution')
plt.show()
```

# Finding correlation between numerical columns and plotting heat map

```python
co=df.select_dtypes(include='number').corr()
co
```

```python
plt.figure(figsize=(15,20))
sns.heatmap(co,annot=True,cmap='YlGnBu')
plt.show()
```
