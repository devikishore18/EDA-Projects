```python
#What is the distribution of bike rentals (cnt) across different hours of the day?
#IMPORTING ALL ESSENTIAL LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("hour.csv")
df1
```

```python
df=pd.DataFrame(df1)
df.head()
```

```python
#What is the distribution of bike rentals (cnt) across different hours of the day?
a=df.groupby('hr')['cnt'].mean().reset_index()
a
```

```python
sns.lineplot(x='hr',y='cnt',data=a,marker='o', linewidth=2)
plt.title('Average Bike Rentals by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Average Rental Count')
plt.xticks(range(0, 24))
plt.grid(True)
```

```python
#Do weekends and weekdays show different rental patterns?
df.columns
```

```python
df['weekday'].value_counts()
```

```python
hourly_grouped = df.groupby(['hr', 'workingday'])['cnt'].mean().reset_index()
hourly_grouped
```

```python
hourly_grouped['day_type'] = hourly_grouped['workingday'].map({0: 'Non-Working Day', 1: 'Working Day'})
hourly_grouped['day_type']
```

```python
sns.lineplot(data=hourly_grouped, x='hr', y='cnt', hue='day_type', marker='o')
plt.title('Average Bike Rentals by Hour (Working vs Non-Working Days)')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Rental Count')
plt.xticks(range(0, 24))
plt.grid(True)
plt.legend(title='Day Type')
plt.tight_layout()
plt.show()
```

```python
#How does bike rental usage vary across seasons (Spring, Summer, Fall, Winter)?
df.columns
```

```python
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
df['season_name'] = df['season'].map(season_map)
hour_season_grouped = df.groupby(['hr', 'season_name'])['cnt'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=hour_season_grouped, x='hr', y='cnt', hue='season_name', marker='o')
plt.title('Average Bike Rentals by Hour Across Seasons')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Rental Count')
plt.xticks(range(0, 24))
plt.grid(True)
plt.legend(title='Season')
plt.tight_layout()
plt.show
```

```python
'''Peak demand hours (e.g., 8â€“9 AM & 5â€“6 PM on weekdays).
Leisure patterns (late morning/afternoon spikes on weekends).
How weekday commuting vs weekend usage differ hour-by-hour.'''
weekday_map = {0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday',
               4: 'Thursday', 5: 'Friday', 6: 'Saturday'}
df['weekday_name'] = df['weekday'].map(weekday_map)
heatmap_data = df.pivot_table(index='hr', columns='weekday_name', values='cnt', aggfunc='mean')
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5, annot=False)
plt.title('Average Hourly Bike Rentals by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Hour of the Day')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
hourly_usage = df.groupby('hr')[['casual', 'registered']].mean().reset_index()
plt.figure(figsize=(12, 6))
plt.bar(hourly_usage['hr'], hourly_usage['casual'], label='Casual', color='#66b3ff')
plt.bar(hourly_usage['hr'], hourly_usage['registered'], bottom=hourly_usage['casual'], label='Registered', color='#ff9999')

plt.title('Average Casual vs Registered Users by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Number of Users')
plt.xticks(range(0, 24))
plt.legend()
plt.tight_layout()
plt.show()
'''Casual users typically rent bikes later in the day and on weekends.

Registered users peak during morning and evening commutes.

This split is valuable for targeted service improvements.'''
```

```python
#ðŸ“Š 1. Casual vs Registered Users by Hour (Average)
hourly_avg = df.groupby('hr')[['casual', 'registered']].mean().reset_index()
plt.figure(figsize=(12, 6))
plt.bar(hourly_avg['hr'], hourly_avg['casual'], label='Casual', color='#66b3ff')
plt.bar(hourly_avg['hr'], hourly_avg['registered'], bottom=hourly_avg['casual'], label='Registered', color='#ff9999')
plt.title('Average Casual vs Registered Users by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Number of Users')
plt.xticks(range(0, 24))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df['day_type'] = df['workingday'].map({1: 'Working Day', 0: 'Weekend'})
hourly_split = df.groupby(['hr', 'day_type'])[['casual', 'registered']].mean().reset_index()
import seaborn as sns
g = sns.catplot(
    data=hourly_split, kind="bar",
    x="hr", y="casual", hue="day_type", ci=None,
    palette="Blues", height=5, aspect=2, label="Casual"
)
g.map_dataframe(sns.barplot, x="hr", y="registered", hue="day_type", data=hourly_split, ci=None, palette="Reds", alpha=0.6)
plt.title('Average Casual vs Registered Users by Hour (Weekday vs Weekend)')
plt.xlabel('Hour')
plt.ylabel('Average Users')
plt.tight_layout()
plt.legend(title='Day Type')
plt.show()
```

```python
#How are temperature (temp) and "feels like" temperature (atemp) correlated?
correlation = df['temp'].corr(df['atemp'])
plt.figure(figsize=(6, 6))
sns.scatterplot(x='temp', y='atemp', data=df, alpha=0.5, color='orange')
plt.title('Correlation Between Temperature and Feels-Like Temperature')
plt.xlabel('Temperature (normalized)')
plt.ylabel('Feels Like Temperature (normalized)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
correlation
```

```python
#How does weather condition (weathersit) affect bike rental volume?
weather_labels = {
    1: "Clear",
    2: "Mist/Cloudy",
    3: "Light Snow/Rain",
    4: "Heavy Rain/Ice"
}
df['weather_desc'] = df['weathersit'].map(weather_labels)
plt.figure(figsize=(8, 5))
sns.boxplot(x='weather_desc', y='cnt', data=df, palette='coolwarm')
plt.title('Bike Rental Volume by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Rental Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Mean rentals per condition
weather_avg = df.groupby('weather_desc')['cnt'].mean().sort_values(ascending=False)
print("Average bike rentals per weather condition:\n")
print(weather_avg)
```

```python
humidity_median = df['hum'].median()
windspeed_median = df['windspeed'].median()
df['humidity_level'] = ['High' if h >= humidity_median else 'Low' for h in df['hum']]
df['windspeed_level'] = ['High' if w >= windspeed_median else 'Low' for w in df['windspeed']]
plt.figure(figsize=(8, 5))
sns.boxplot(x='humidity_level', y='cnt', data=df, palette='coolwarm')
plt.title('Bike Rentals vs Humidity Level')
plt.xlabel('Humidity Level')
plt.ylabel('Rental Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Boxplot for Windspeed
plt.figure(figsize=(8, 5))
sns.boxplot(x='windspeed_level', y='cnt', data=df, palette='Set2')
plt.title('Bike Rentals vs Windspeed Level')
plt.xlabel('Windspeed Level')
plt.ylabel('Rental Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#What is the effect of rain, snow, or fog (weather conditions) on hourly demand?
hour_weather = df.groupby(['hr', 'weather_desc'])['cnt'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=hour_weather, x='hr', y='cnt', hue='weather_desc', marker='o')
plt.title('Hourly Bike Rental Demand by Weather Condition')
plt.xlabel('Hour of Day')
plt.ylabel('Average Rental Count')
plt.legend(title='Weather Condition')
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
#Are there any noticeable trends in usage between 2011 and 2012 (yr)?
df['year'] = df['yr'].map({0: '2011', 1: '2012'})
monthly_usage = df.groupby(['year', 'mnth'])['cnt'].sum().reset_index()
month_map = {
    1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun',
    7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'
}
monthly_usage['month_name'] = monthly_usage['mnth'].map(month_map)
monthly_usage['month_name'] = pd.Categorical(monthly_usage['month_name'],
                                              categories=list(month_map.values()), ordered=True)
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_usage, x='month_name', y='cnt', hue='year', marker='o')
plt.title('Monthly Bike Rentals in 2011 vs 2012')
plt.xlabel('Month')
plt.ylabel('Total Rentals')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Is there a correlation between the hour of the day and rental count that differs by season?
correlations = df.groupby('season_name').apply(lambda x: x['hr'].corr(x['cnt']))
print("Correlation between hour and rental count per season:\n")
print(correlations)
# Plot rental trends by hour and season
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='hr', y='cnt', hue='season_name', estimator='mean', marker='o')
plt.title('Average Rental Count by Hour in Each Season')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Rentals')
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```
